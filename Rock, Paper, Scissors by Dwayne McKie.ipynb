```python
#import the library that contains the built-in function randint()
import random as rand
```

```python

#greet user to the game
print("Welcome to the Rock, Paper, Scissors Game!!!")

#create variable to store rock, paper or scissors string
computer_thing = " "

#create variable that will store value for continuation of game by user (or not)
answer = 'y'



#create loop that will allow user to play game until the user decides not to play any longer
while answer != 'n' :

#prompt the user to enter rock, paper, or scissors
#store selection into a string variable
    user_thing = input("Enter rock, paper, or scissors:    ")
    user_thing = user_thing.lower()

#print user selection    
    print(" ")
    print("User selects", user_thing)

#create random integer (1, 2, or 3) that will decide the object chosen by the computer
    computer_thing_random_number = rand.randint(1, 3)

#convert random integer selection to a string value or rock, paper or scissors    
    if computer_thing_random_number == 1 :
        computer_thing = "rock"
    elif computer_thing_random_number == 2 :
        computer_thing = "paper"
    else :
        computer_thing = "scissors"

#print computer selection        
    print("Computer selects", computer_thing)    
    
#compare selections if user enters rock and the outcome
    if user_thing == "rock" :
        if computer_thing == "rock" :
            print("rock & rock are the same. It's a draw.")
        elif computer_thing == "paper" :
            print("paper covers rock. You lose.")
        else :
            print("rock smashes scissors. You win!!!!")
        
#compare selections if user enters paper and the outcome        
    if user_thing == "paper" :
        if computer_thing == "rock" :
            print("paper covers rock. You win!!!")
        elif computer_thing == "paper" :
            print("paper & paper are the same. It's a draw.")
        else :
            print("scissors cuts paper. You lose.")

#compare selections if user enters scissors and the outcome            
    if user_thing == "scissors" :
        if computer_thing == "rock" :
            print("rock smashes scissors. You lose.")
        elif computer_thing == "paper" :
            print("scissors cuts paper. You win!!!")
        else :
            print("scissors & scissors are the same. It's a draw.")

#ask user if the user wants to continue playing game            
    answer = input("Do you want to continue? y/n:    ")        

#if user leaves game, thank the user for playing    
print(" ")
print(" ")
print("Thank you for playing Rock, Paper, Scissors!!!")
```
